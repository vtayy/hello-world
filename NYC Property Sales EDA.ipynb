Since we know that Building Class Category have its own identification code in BUILDING CLASS AS OF FINAL ROLL, we can remove the BUILDING CLASS CATEGORY column for a simplier table. LOT represents the property's unique location and BLOCK is another identification code used for Department of Finance to distinguish the different units of apartment from another, such as different condo in a single building. However, I don't think they are essential in this EDA process, so I will also take those two columns out. 

NTA is a more specific location of NEIGHBORHOOD.

```python
columns_to_drop = ['BUILDING CLASS CATEGORY', 'APARTMENT NUMBER', 'Community Board', 'Council District',
                   'Census Tract', 'BIN', 'BBL', 'BLOCK', 'LOT', 'EASE-MENT', 'GROSS SQUARE FEET', 'COMMERCIAL UNITS',
                   'TOTAL UNITS']
def drop_cols(sales):
    return sales.drop(columns=columns_to_drop)
sales.pipe(drop_cols)
```
